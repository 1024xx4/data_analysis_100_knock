Node（頂点）間の Link の太さを変えていく（重みづけを行う）ことで、物流の最適 Route をわかりやすく可視化していく。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# Data 読み込み
df_w = pd.read_csv('sample_code/chapter_6/network_weight.csv')
df_p = pd.read_csv('sample_code/chapter_6/network_pos.csv')

# Edge の重みの List 化
size = 10
edge_weight = []
for i in range(len(df_w)):
    for j in range(len(df_w.columns)):
        edge_weight.append(df_w.iloc[i][j] * size)

# Class Object の作成
G = nx.Graph()

# 頂点の設定
for i in range(len(df_w.columns)):
    G.add_node(df_w.columns[i])

# 辺の設定
for i in range(len(df_w.columns)):
    for j in range(len(df_w.columns)):
        G.add_edge(df_w.columns[i], df_w.columns[j])

# 座標の設定
pos = {}
for i in range(len(df_w.columns)):
    node = df_w.columns[i]
    pos[node] = (df_p[node][0], df_p[node][1])

# 描画
nx.draw(G, pos, with_labels=True, font_size=16, node_size=1000, node_color='k', font_color='w', width=edge_weight)

# 表示
plt.show()
```
